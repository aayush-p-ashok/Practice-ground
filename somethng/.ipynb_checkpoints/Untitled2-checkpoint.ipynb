This is an animated version of E-3.1

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Necessary for 3D plotting
from matplotlib.animation import FuncAnimation

# Define a figure and a fixed set of axes that we'll be reusing
fig = plt.figure(figsize=(15, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)

# Original potential calculation function (can be reused without changes)
def potential(q, x, y, x0, y0):
    k = 9e9
    r = np.sqrt((x - x0) ** 2 + (y - y0) ** 2 + 1e-9)
    return k * q / r

# Animation update function for each frame
def update(y_slice):
    ax1.cla()  # Clear previous 3D plot
    ax2.cla()  # Clear previous 2D plot

    # Define parameters
    elev = 25
    azim = 45
    q = 1
    s = 1.5
    n = 12

    # Set up for potential plot
    theta0 = np.pi / n
    R = s / (2 * np.sin(theta0))

    # 2D grid for calculation
    X = np.linspace(-5, 5, 500)
    Y = np.linspace(-5, 5, 500)
    x, y = np.meshgrid(X, Y)

    # Calculate vertices positions
    Xm = np.array([R * np.cos(2 * m * theta0) for m in range(n)])
    Ym = np.array([R * np.sin(2 * m * theta0) for m in range(n)])

    # Calculate net potential
    Vnet = sum(potential(q, x, y, xm, ym) for xm, ym in zip(Xm, Ym))

    epsilon = 1e-7
    Vplot = np.sign(Vnet) * np.log10(np.abs(Vnet) + epsilon)

    # Plot the 3D surface
    surf = ax1.plot_surface(x, y, Vplot, cmap='magma', edgecolor='none',
                            antialiased=True, rcount=100, ccount=100, alpha=0.9)
    
    # Zero potential contours
    ax1.contour(x, y, Vplot, levels=[0], colors='green', linewidths=2, zdir='z', offset=Vnet.min())
    ax1.contour3D(x, y, Vplot, levels=[0], colors='#00FFFF', linewidths=2)
    
    # Set 3D plot properties
    ax1.set_xlabel('x', labelpad=10)
    ax1.set_ylabel('y', labelpad=10)
    ax1.set_zlabel(r'$\log(V)$', labelpad=10)
    ax1.set_title('3D Potential', pad=20, y=1.05)
    ax1.view_init(elev=elev, azim=azim)
    ax1.dist = 8
    ax1.set_box_aspect([1, 1, 0.8])
    ax1.set_xlim(-6, 6)
    ax1.set_ylim(-6, 6)

    # Create slice plane
    slice_x = np.linspace(-6, 6, 100)
    slice_z = np.linspace(Vplot.min(), Vplot.max(), 100)
    slice_X, slice_Z = np.meshgrid(slice_x, slice_z)
    slice_Y = np.full_like(slice_X, y_slice)  # Y-value varies over frames

    # Plot the transparent plane
    ax1.plot_surface(slice_X, slice_Y, slice_Z, 
                    alpha=0.2, color='yellow', shade=False)  # Plane color, no shading
    
    # Optional: Add a line at the intersection
    y_idx = np.abs(Y - y_slice).argmin()  # Find index closest to y_slice
    ax1.plot(X, [y_slice]*len(X), Vplot[y_idx, :], color='yellow', linewidth=2)
    
    # Plot the 2D slice
    ax2.plot(X, Vplot[y_idx, :], 'b-', linewidth=2)
    ax2.grid(True)
    ax2.set_xlabel('x')
    ax2.set_ylabel(r'$\log(V)$')
    ax2.set_title(f'Potential at y = {y_slice:.2f}')
    ax2.set_xlim(-6, 6)

    # Add horizontal line at V=0
    ax2.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    
    # Add vertical lines where charges are located
    for xm, ym in zip(Xm, Ym):
        if abs(ym - y_slice) < 0.2:  # Only show charges near the slice
            ax2.axvline(x=xm, color='g', linestyle=':', alpha=0.5)

    # Adjust layout for tight display
    fig.tight_layout()
    plt.show()

# Range of y_slice values to animate over
y_slices = np.linspace(-5, 5, 100)  # Smooth transition

# Set up animation
ani = FuncAnimation(fig, update, frames=y_slices, blit=False, repeat=True)

# To display or save, uncomment one of these lines:
# ani.save('potential_animation.mp4', fps=20, dpi=80)

```
